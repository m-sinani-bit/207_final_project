```python
data = pd.read_csv('/content/drive/MyDrive/207Final/207HyperparameterAugmentationData.csv')
```

```python
data['rank_training_loss'] = data['training_loss'].rank(ascending=True)

data['rank_validation_loss'] = data['validation_loss'].rank(ascending=True)

data['rank_precision'] = data['precision'].rank(ascending=False)

data['rank_accuracy'] = data['accuracy'].rank(ascending=False)
```

```python
data.head()
```

```python
data['sum_of_ranks'] = data[['rank_training_loss', 'rank_validation_loss', 'rank_precision', 'rank_accuracy']].sum(axis=1)
```

```python
sorted_data = data.sort_values(by='sum_of_ranks', ascending=True)

# Display the top 5 rows with the lowest values in 'sum_of_ranks' column
sorted_data.head(10)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sorted_data = data.sort_values(by='sum_of_ranks', ascending=True)

# Get the top 5 rows with the lowest values in 'sum_of_ranks' column
top_5_lowest = sorted_data.head(5)

# Create a heatmap-like table using Seaborn
plt.figure(figsize=(10, 4))  # Adjust the figure size as needed
sns.set(style="whitegrid")   # Set the style
sns.heatmap(top_5_lowest[['column1', 'column2', 'sum_of_ranks']], annot=True, cmap='coolwarm', cbar=False)
plt.title('Top 5 Rows with Lowest Sum of Ranks')
plt.show()
```
